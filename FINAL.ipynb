```python
'''
The next factor I want to look at is the idea of killing the opponents character. 
Killing the opponents character gives the player about 300 gold (used for buying items and becoming more powerful),
a large amount of experience, and perhaps most importantly, a large mental boost for the rest of the game. [awk]
As such, many players empahsize killing their lane opponent vesus the other objectives in the game. 
In this section, we will break down the average kills per won game and see if having a high kill amount plays a signifigant
role in game wins. 
'''
from scipy import stats

def get_kills(match):
    return match.kills

def model_summary(X,y):
    #credit goes to https://stackoverflow.com/questions/27928275/find-p-value-significance-in-scikit-learn-linearregression
    #for this function.
    lm = LogisticRegression()
    lm.fit(X,y)
    params = np.append(lm.intercept_,lm.coef_)
    predictions = lm.predict(X)

    newX = pd.DataFrame({"Constant":np.ones(len(X))}).join(pd.DataFrame(X))
    MSE = (sum((y-predictions)**2))/(len(newX)-len(newX.columns))

    # Note if you don't want to use a DataFrame replace the two lines above with
    # newX = np.append(np.ones((len(X),1)), X, axis=1)
    # MSE = (sum((y-predictions)**2))/(len(newX)-len(newX[0]))

    var_b = MSE*(np.linalg.inv(np.dot(newX.T,newX)).diagonal())
    sd_b = np.sqrt(var_b)
    ts_b = params/ sd_b

    p_values =[2*(1-stats.t.cdf(np.abs(i),(len(newX)-1))) for i in ts_b]

    sd_b = np.round(sd_b,3)
    ts_b = np.round(ts_b,3)
    p_values = np.round(p_values,3)
    params = np.round(params,4)

    myDF3 = pd.DataFrame()
    myDF3["Coefficients"],myDF3["Standard Errors"],myDF3["t values"],myDF3["Probabilites"] = [params,sd_b,ts_b,p_values]
    print(myDF3)

    
won_matches = match_df[match_df['win'] == 1]
kills = won_matches['kills']
print(kills.quantile([.25,.50,.75]))



'''
For most won games, the median amount of kills is only around 38! 
Lets split the games into two catagories now, one with 'above average' kills and one with 'below average' kills,
then run a chisq test to see if getting above the average amount of kills or below the average amount is statistically
signifigant to winning a game!
'''



print(model_summary(match_df[['kills']].astype(float), match_df['win'].astype(float)))

'''
#TODO: explain why this model supports the idea that kills are important (p_val small)
'''

#TODO: plot the distribution of kills
kill_subset = match_df[['kills','win']]

wr_above_avg_kills = kill_subset[kill_subset['kills'] > 30].groupby("win").count()
print("wr_above_avg_kills .25\n")
print(wr_above_avg_kills)
print(sp.chisquare(wr_above_avg_kills)[1] < .05)

wr_below_avg_kills = kill_subset[kill_subset['kills'] <= 30].groupby("win").count()
print("wr_below_avg_kills .25")
print(wr_below_avg_kills)
print(sp.chisquare(wr_below_avg_kills)[1] < .05)




wr_above_avg_kills = kill_subset[kill_subset['kills'] > 38].groupby("win").count()
print("wr_above_avg_kills .5\n")
print(wr_above_avg_kills)
print(sp.chisquare(wr_above_avg_kills)[1] < .05)

wr_below_avg_kills = kill_subset[kill_subset['kills'] <= 38].groupby("win").count()
print("wr_below_avg_kills .5")
print(wr_below_avg_kills)
print(sp.chisquare(wr_below_avg_kills)[1] < .05)


wr_above_avg_kills = kill_subset[kill_subset['kills'] > 46].groupby("win").count()
print("wr_above_avg_kills .75\n")
print(wr_above_avg_kills)
print(sp.chisquare(wr_above_avg_kills)[1] < .05)

wr_below_avg_kills = kill_subset[kill_subset['kills'] <= 46].groupby("win").count()
print("wr_below_avg_kills .75")
print(wr_below_avg_kills)
print(sp.chisquare(wr_below_avg_kills)[1] < .05)



'''
TODO: explain this shit.
TODO: maybe test the different quantiles
TODO: check using logistic regression as well
'''

# '''
# TODO: more explaination
# '''
```

```python
'''
##TODO: Plot the distribution of players playing each role

In League of Legends, players are assigned a 'lane' (Top, Jungle, Middle, ADC, and Support) and can bring a single 
character to their lane. 
Each character in League of Legends has a 'role' assigned to them. 

For Example: 
Aatrox (https://universe.leagueoflegends.com/en_US/champion/aatrox/) would be considered a Juggernaut - someone who
deals a lot of damage, can survive for a long time, but doesnt have a lot of mobility or movement skills. 

In this part, we will be looking at each lane's role and see whether or not the presence of a certain role (or the lack of presence)
affects the win rate.
'''

## TOP LANE
print("TOP: ")
top_df = match_df[['Top1', 'win']]
top_roles = top_df['Top1'].unique()
print(top_roles) #the number of unique roles people play in the top lane
print("\n")
    
'''
From a curosary glance it looks like Juggernauts have the highest winrate while any kind of 
magic character (Battlemage, Enchanter, Artillery) have the lowest winrates 

Lets see which roles are statistically signifigant using a chi-sqaure test
'''

for role in top_roles:
    print("\n")
    print("ROLE: ", role)
    tbl = top_df[top_df['Top1'] == role].groupby("win").count()
    print(tbl)
    print("WR: ", tbl.loc[1] / (tbl.loc[0] + tbl.loc[1]))
    print("chisq: ",sp.chisquare(tbl)[0])
    print("p_value: ", sp.chisquare(tbl)[1])
    print(sp.chisquare(tbl)[1] < .05)
    
'''
Thus as it turns out, for the top lane, the only statistically signifigant roles are 
Warden and Catcher. The relationship is in the positive direction, as having these roles on your 
team make you more likely to win the game. 

Personal experience: 
    As someone whose played this game a lot, these results make sense. Generally wardens have been the prefered 
    choice both in professional play and in casual play due to their reliability. I did not expect 
    catchers to be in the list however, so its very interesting to see that they are signifigant. 
'''


```
